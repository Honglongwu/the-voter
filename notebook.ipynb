```python
%matplotlib inline
from Main import *
import matplotlib.pyplot as plt

df = create_posts_df()
names = df.candidate.unique()

```

```python
df.groupby('candidate').count().post.plot(kind='bar').set_ylabel('number of posts')
```

```python
df['dateStr'] = df.date.apply(lambda x: x.strftime('%Y-%m'))
ax = df.groupby(['candidate','dateStr']).count().post.unstack(level=0).plot(marker='.')
ax.legend(bbox_to_anchor=(1.42, 1.02))
ax.set_ylabel('number of posts')
ax.set_xlabel('month')
```
