```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import classification_report
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
import hunspell

#an Hebrew spell checker and a stemmer
hspell = hunspell.HunSpell('nlp/he.dic', 'nlp/he.aff')
def my_stemmer(word):
    hstem = hspell.stem(word.encode('utf8'))
    if len(hstem)!=1:
        return word
    else:
        stem = hstem[0].decode('utf8')
        if u'ה'+stem==word:
            return word
        return stem

#define a tokenizer using Yoav Goldbergs' tokenizer http://www.cs.bgu.ac.il/~yoavg/software/hebtokenizer/ and hspell/hunspell stemmer
#skip non words (e.g, punctuations)
def my_tokenizer(sent):
    return [my_stemmer(w) for t,w in tokenizer.tokenize(sent) if t=='HEB' and len(w)>1]


#Hebrew stop words.
STOP_WORDS = set([w.decode('utf8').strip() for w in open('nlp/stopwords.txt','r') if len(w.strip())>0])

#train size is 90% of the data. 10% is left for testing
TRAIN_SIZE = int(0.9 * len(df))
    
#create train and test sets
X = df.post.values
Y = df.candidate.values
X_train = X[:TRAIN_SIZE]
X_test  = X[TRAIN_SIZE:]
Y_train = list(Y[:TRAIN_SIZE])
Y_test  = list(Y[TRAIN_SIZE:])


#run this method to find the best parameters of the mode. you can add other parameters, but it will cost you some time
def find_best_params():
    #create a tfidf vectorizer. we don't have stemmer for hebrew :(
    vectorizer = TfidfVectorizer(stop_words=STOP_WORDS,tokenizer=my_tokenizer)

    #select K best features using chi-squared statistic for each class/feature combination
    ch2 = SelectKBest(chi2)

    #linear SVM training. we will use this classifier as it allows probabilties estimates. you can try other classifiers...
    clf = SVC(kernel='linear')

    #create a pipeline of tfidf transformation, selection of the k best features and the acctual classifier
    pl_clf = Pipeline([('vectorizer', vectorizer),('kbest',ch2),('clf', clf)])

    #different parameters in the pieline that we are going to tune
    parameters = {'vectorizer__ngram_range': [(1,3)],
                  'vectorizer__max_df': [0.3,1.0],
                  'vectorizer__min_df': [2,5],
                  'kbest__k': [6000,'all']}

    #fit the model and find the best parameters using cross validation. this is going to take a while...
    gs_clf = GridSearchCV(pl_clf, parameters)
    gs_clf = gs_clf.fit(X_train, Y_train)

    #print the best parameters values
    best_parameters, score, _ = max(gs_clf.grid_scores_, key=lambda x: x[1])
    for param_name in sorted(parameters.keys()):
        print("%s: %r" % (param_name, best_parameters[param_name]))




    
```

```python
import pickle

#create a new classifier with the best values that we computed above. since kbest value is 'all', we don't need SelectKBest 
vectorizer = TfidfVectorizer(stop_words=STOP_WORDS,tokenizer=my_tokenizer,ngram_range=(1,3),sublinear_tf=True,min_df=2,max_df=1.0)
clf = SVC(kernel='linear',probability=True) #enable probabilities estimates
pl_clf = Pipeline([('vectorizer', vectorizer),('clf', clf)])
pl_clf.fit(X_train,Y_train)

#save the model
pickle.dump(pl_clf,open( "pl_clf.pkl", "w" ))

#evaluation on the test set
predicted = pl_clf.predict(X_test)
print(classification_report(Y_test, predicted))
```
